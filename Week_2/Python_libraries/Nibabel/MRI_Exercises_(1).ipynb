# MRI Visualization and Normalization Exercises

Synthetic MRI volume provided with this notebook.


```python
import numpy as np
import nibabel as nib
import matplotlib.pyplot as plt
```

```python
nii = nib.load('sample_mri.nii.gz')
volume = nii.get_fdata()
volume.shape
```

```python
def plot_slices(volume, indices):
    plt.figure(figsize=(15,5))
    for i, idx in enumerate(indices):
        plt.subplot(1, len(indices), i+1)
        plt.imshow(volume[:, :, idx], cmap='gray')
        plt.title(f'Slice {idx}')
        plt.axis('off')
    plt.show()
```

```python
depth = volume.shape[2]
indices = np.linspace(depth*0.3, depth*0.7, 5).astype(int)
plot_slices(volume, indices)
```

```python
def normalize_volume(volume):
    return (volume - volume.min()) / (volume.max() - volume.min())
```

```python
norm_volume = normalize_volume(volume)
norm_volume.min(), norm_volume.max()
```

```python
plot_slices(norm_volume, indices)
```
